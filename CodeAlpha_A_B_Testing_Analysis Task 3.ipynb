```python
new_column_names = [
    'Campaign Name', 'Date', 'Amount Spent', 'Impressions', 'Reach',
    'Number of Clicks', 'Number of Searches', 'Number of views',
    'Number Added to cart', 'Purchase Number'
]

test_df.columns = new_column_names

test_df.head()
```

```python
columns_to_fill = [
    'Reach', 'Impressions', 'Number of Clicks', 'Number of Searches',
    'Number of views', 'Number Added to cart', 'Purchase Number'
]

for column in columns_to_fill:
    mean_value = control_df[column].mean()
    control_df[column].fillna(mean_value, inplace=True)

control_df.head()
```

```python
from scipy.stats import shapiro

shapiro_results_list = []

test_stat_control, p_value_control = shapiro(control_df['Purchase Number'])

shapiro_results_list.append({'Group': 'Control', 'Test Statistic': test_stat_control, 'P-value': p_value_control})

test_stat_test, p_value_test = shapiro(test_df['Purchase Number'])

shapiro_results_list.append({'Group': 'Test', 'Test Statistic': test_stat_test, 'P-value': p_value_test})

shapiro_results = pd.DataFrame(shapiro_results_list)

shapiro_results
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(control_df['Purchase Number'], test_df['Purchase Number'])

t_stat, p_value
```

```python
control_df['Cost per Conversion'] = control_df['Amount Spent'] / control_df['Purchase Number']
test_df['Cost per Conversion'] = test_df['Amount Spent'] / test_df['Purchase Number']

average_cost_control = control_df['Cost per Conversion'].mean()
average_cost_test = test_df['Cost per Conversion'].mean()

average_cost_control, average_cost_test
```

```python
merged = pd.concat([control_df, test_df], ignore_index=True)

merged.head()
```

```python
merged['CTR'] = (merged['Number of Clicks'] / merged['Impressions']) * 100

merged['Conversion Rate'] = (merged['Purchase Number'] / merged['Number of Clicks']) * 100

merged['CPC'] = merged['Amount Spent'] / merged['Number of Clicks']

merged['ROI'] = ((merged['Purchase Number'] - merged['Amount Spent']) / merged['Amount Spent']) * 100

merged.head()
```

```python
metrics_to_test = ['CTR', 'Conversion Rate', 'CPC']

results_list = []

for metric in metrics_to_test:

    t_stat, p_value = ttest_ind(merged.loc[merged['Campaign Name'] == 'Control Campaign', metric],
                                merged.loc[merged['Campaign Name'] == 'Test Campaign', metric])

    metric_results = pd.DataFrame({'Metric': [metric],
                                   'T-Statistic': [t_stat],
                                   'P-Value': [p_value]})

    results_list.append(metric_results)

t_test_results = pd.concat(results_list, ignore_index=True)

t_test_results
```
